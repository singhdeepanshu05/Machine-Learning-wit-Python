## Q3. What is the idea behind the Gradient Descent?

Ans : The idea behind Gradient descent is to minimize the error and to reach as close as to global minima.

## Q4. What is the difference between stochastic gradient descent and gradient descent?

Gradient descent gives value where error is less but it takes more time as it has to through a number of iteration or epochs, where as in stochastic gradient descent the value of x and y is picked at random which will give us a fast result but the error may be large as there are no steps or epochs.

## Q5. What is Learning rate ? Which type of learning rate you will prefer and why ?

Learning rate is the criteria through which system learns how much steps it should take to reach closer to global minima, a prefer learning rate would be the one which converges in slow pace and does not diverger values.
